# Chip characterization

**Here is Chip characterization of my sample table 3.1 and fig. 3.7. Where I fitted experimental data critical current to to fit with an exponential function**

```python
#D-series  Josephson Junctions D19-D30
import math as mt
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.ticker as tck
import matplotlib.font_manager as fm




x=[455.0,   455.0,  505.0,   505.0,     556.0,    556.0,   606.0,    606.0,  657.0,   657.0,    707.0,  707.0]
y=[1.86E+04, 1.58E+04,  1.37E+04,  8.15E+03,  1.31E+03,  4.52E+03,  1.67E+03, 7.14E+02, 2.14E+02, 9.52E+02, 3.57E+02, 3.57E+02]

xx= np.array(x)
yy= np.array(y)
p = np.polyfit(xx, np.log(yy), 1)

#area of bjj = w*t= 2.8 * 10^-6 * 30 * 10^-9= 84 *10^-15 m^2 = 84*10^-11 cm^2
#I_c=[15.6, 13.2,  11.5,  6.8,  1,  3.7,  1.4, 0.59, 0.17, 0.79 , 0.29, 0.3]µA



a = np.exp(p[1])
b = p[0]

y_fitted = a * np.exp(b * xx) 
J_c0 = print(a)
D_0 = print(b)






fig, ax = plt.subplots(1, 1, figsize=(9, 6), sharex=True, sharey=True)#
#ax.plot(x,r, color = 'red')
size = [100,500,100,500]
ax.plot(xx, y_fitted, 'k', label='Fitted curve')
ax.scatter(xx, yy, label='Raw data')
ax.set_title("Evaluation of D19-D30")
plt.xlabel('Dose', fontsize=20)
plt.ylabel('j$_c$($KA/cm^2)$', fontsize=20)
plt.yscale("log")



plt.legend(loc="upper right")
plt.rcParams['figure.figsize'] = [9, 6]
plt.xlabel("V$(mV)$", fontname="serif", style='italic', fontsize=18)  #fontname="Comic Sans MS"
plt.ylabel("$j_c$($KA/cm^2$)", fontname="serif", style='italic', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


#ax.yaxis.set_minor_locator(tck.AutoMinorLocator(5))
ax.xaxis.set_minor_locator(tck.AutoMinorLocator(2))
ax.tick_params(axis='x',which='minor', length=4, width=1.1, direction="in") # direction="in"
ax.tick_params(axis='x', which='major', length=6, width=2.1, direction='in')
ax.tick_params(axis='y',which='minor',  length=4, width=1.1, direction='in')
ax.tick_params(axis='y', which='major', length=6, width=2.1, direction='in')
ax.xaxis.set_ticks_position('default')
ax.yaxis.set_ticks_position('default')
ax.tick_params(bottom=True, top=True, left=True, right=True)






```
